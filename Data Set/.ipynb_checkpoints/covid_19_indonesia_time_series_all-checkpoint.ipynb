```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid_19_indonesia_time_series_all.csv")
df.head()
```

```python
to_drop = ['Date',
          'Location ISO Code',
          'Latitude','New Cases per Million',
          'Total Cases per Million',
          'New Deaths per Million',
          'Total Deaths per Million',
          'Total Deaths per 100rb',
           'Case Fatality Rate',
          'Case Recovered Rate',
          'Growth Factor of New Cases',
          'Growth Factor of New Deaths',
          'Longitude',
          'Population',
          'Population Density',
          'Area (km2)',
          'Total Rural Villages',
          'Total Urban Villages',
          'Total Districts',
          'Total Cities',
          'Total Regencies',
          'Special Status',
          'Location Level',
          'Time Zone',
          'Continent',
          'Country',
           'City or Regency',
          'Province','Island']
```

```python
df.drop(to_drop, inplace=True, axis=1)
```

```python
df.info()
```

```python
x = df.drop(["Location"], axis=1)
x.head(11)
```

```python
y = df["Location"]
y.head(11)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

modelnb = GaussianNB()
```

```python
nbtrain = modelnb.fit(x, y)
df.head(11)
```

```python
x_test = df.drop(["Location"], axis=1)
x_test.head(11)
```

```python
y_uji = df["Location"]
y_uji.head(11)
```

```python
Y_predict = nbtrain.predict(x_test)
print("Prediksi Naive Bayes : ",Y_predict)
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_uji, Y_predict)
print("Akurasi Naive Bayes : ",accuracy)
```

```python
# Menghitung nilai akurasi dari klasifikasi naive bayes 
from sklearn.metrics import classification_report
print(classification_report(y_uji, Y_predict))
```
