```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Rasio Bed To Population Rumah Sakit.csv")
df.head(10)
```

```python
df.info()
```

```python
x = df.drop(["Provinsi"], axis=1)
x.head(11)
```

```python
y = df["Provinsi"]
y.head(11)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

modelnb = GaussianNB()
```

```python
nbtrain = modelnb.fit(x, y)
df.head(11)
```

```python
x_test = df.drop(["Provinsi"], axis=1)
x_test.head(11)
```

```python
y_uji = df["Provinsi"]
y_uji.head(11)
```

```python
Y_predict = nbtrain.predict(x_test)
print("Prediksi Naive Bayes : ",Y_predict)
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_uji, Y_predict)
print("Akurasi Naive Bayes : ",accuracy)
```

```python
# Menghitung nilai akurasi dari klasifikasi naive bayes 
from sklearn.metrics import classification_report
print(classification_report(y_uji, Y_predict))
```
